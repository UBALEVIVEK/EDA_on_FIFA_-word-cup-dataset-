```python
# make a list of number and variable from value calumn 
set_char = set()
def add_char(X):
    for i in X:
        set_char.add(i)
    return True
df["Value"].apply(add_char)
set_char
```

Trying to create a set of unique characters from the values in the "Value" column of a DataFrame named df. The add_char function is applied to each element in the "Value" column using the apply method.

```python
# replace ‚Ç¨,M,K, with space,number*1000000,number*1000 respactively
def get_value(x):
    x= x.replace('‚Ç¨','')
    if 'M' in x:
        x=x.replace('M','')
        return float(x)*1000000
    elif 'K' in x:
        x=x.replace('K','')
        return float(x)*1000 
    else:
        return float(x)
df["Value"]=df["Value"].apply(get_value)
```

When you run this code, set_char will contain the unique characters present in the "Value" column of your DataFrame.
